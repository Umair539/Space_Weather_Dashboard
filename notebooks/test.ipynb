```python
import pandas as pd
from sqlalchemy import create_engine

# 1. Point to your database file
engine = create_engine("sqlite:///../data/transformed/noaa_data.db")

# 2. Define your query
query = (
    f"SELECT "
    f"strftime('%Y-%m-%d %H:00:00', time) AS hourly_bucket, "
    f"avg(speed) AS avg_speed "
    f"FROM solar "
    f"GROUP BY hourly_bucket "
    f"ORDER BY hourly_bucket ASC;"
)

# 3. Run and print
df = pd.read_sql(query, engine)
print(df)
```
